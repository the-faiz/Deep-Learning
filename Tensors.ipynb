###Gradient

```python
x = torch.tensor(3.)
w = torch.tensor(4., requires_grad=True)
b = torch.tensor(5., requires_grad=True)
y = w * x + b
y.backward() #Gradient Computation
print('dy/dx:', x.grad)
print('dy/dw:', w.grad)
print('dy/db:', b.grad)
```

###Interconversion with numpy

```python
import numpy as np
x = np.array([[1, 2], [3, 4.]])
y = torch.from_numpy(x) #Conversion to tensor from nupy
z = y.numpy() #Conversion to numpy from tensor
```
